First, we set up the environment necessary for this project (importing required libraries and modules, ensuring that plots display correctly, and reading the `.csv` file we'll be analyzing into `pandas` as a dataframe).

```python
# import pandas and pyplot module of matplotlib
import pandas as pd
import matplotlib.pyplot as plt
# ensure that plots appear inline
%matplotlib inline
# read csv file into pandas
recent_grads = pd.read_csv('recent-grads.csv')
```

Next, we'll do some basic data exploration.

```python
# returns first row of dataframe formatted as a table
print(recent_grads.iloc[0])
# returns first and last rows of dataframe
print(recent_grads.head(1), recent_grads.tail(1))
# generates summary statistics for all numeric columns
recent_grads.describe()
```

Now we'll eliminate rows with missing values. (As the output of the below cell shows, only one row is dropped.)

```python
# calculate number of rows in original dataset
raw_data_count = len(recent_grads)
# drop rows with missing values,
# return result to recent_grads variable
recent_grads = recent_grads.dropna()
# calculate number of rows in cleaned dataset
cleaned_data_count = len(recent_grads)
# compare original and cleaned dataset row counts
print(raw_data_count, cleaned_data_count)
```

## Scatter Plots

In the below cells, we'll generate scatter plots to explore the relationships between various columns in our table.

## Scatter Matrix Plots

Next, we're going to create scatter matrix plots to simultaneously explore relationships between columns and individual column distributions (first importing the `scatter_matrix` function from the `pandas.plotting` module).

```python
from pandas.plotting import scatter_matrix
```

```python
# create scatter matrix plot comparing sample size and median income
ax = scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(7,7))
# set title and create more interpretable, evenly spaced axis labels
plt.suptitle('Scatter Matrix Plot: Sample Size and Median Income')
ax[0,0].set_ylabel('Sample Size', labelpad=11)
ax[1,0].set_ylabel('Median Income')
ax[1,0].set_xlabel('Sample Size', labelpad=14)
ax[1,1].set_xlabel('Median Income')
```

```python
# create scatter matrix plot comparing sample size, median income, and unemployment rate
ax = scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,7))
# set title and create more interpretable, evenly spaced axis labels
plt.suptitle('Scatter Matrix Plot: Sample Size, Median Income, and Unemployment Rate')
ax[0,0].set_ylabel('Sample Size', labelpad=13)
ax[1,0].set_ylabel('Median Income')
ax[2,0].set_ylabel('Unemployment Rate', labelpad=18)
ax[2,0].set_xlabel('Sample Size', labelpad=13)
ax[2,1].set_xlabel('Median Income')
ax[2,2].set_xlabel('Unemployment Rate', labelpad=16)
```

```python
# create scatter matrix plot comparing share of female graduates and median income
ax = scatter_matrix(recent_grads[['ShareWomen', 'Median']], figsize=(7,7))
# set title and create more interpretable, evenly spaced axis labels
plt.suptitle('Scatter Matrix Plot: Proportion of Female Graduates and Median Income')
ax[0,0].set_ylabel('Proportion of Female Graduates', labelpad=20)
ax[1,0].set_ylabel('Median Income')
ax[1,0].set_xlabel('Proportion of Female Graduates', labelpad=22)
ax[1,1].set_xlabel('Median Income')
# rename y-tick labels for ShareWomen histogram to avoid rounding issue
ax[0,0].set_yticklabels(['0.0', '0.2', '0.4', '0.6', '0.8'])
```

```python
# create scatter matrix plot comparing number of full-time workers and median income
ax = scatter_matrix(recent_grads[['Full_time', 'Median']], figsize=(7,7))
# set title and create more interpretable, evenly spaced axis labels
plt.suptitle('Scatter Matrix Plot: Full-Time Workers and Median Income')
ax[0,0].set_ylabel('Respondents Working Full Time', labelpad=3)
ax[1,0].set_ylabel('Median Income')
ax[1,0].set_xlabel('Respondents Working Full Time', labelpad=5)
ax[1,1].set_xlabel('Median Income')
# add note to figure to clarify definition of full-time work
plt.figtext(.25, -.05, "* Full-time work = 35 or more hours per week")
```
